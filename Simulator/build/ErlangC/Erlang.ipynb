# Using the Erlang calculation library

## Loading libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import Erlang
```

## Calculating E\[NQ\] for M/M/1 model

```python
rho=np.linspace(0.5,0.99,100)
ENQ=[];

for r in rho:
    l=r*100
    mu=100
    c=1
    ENQ.append(Erlang.ErlangC_ENQ(l,mu,c))

plt.plot(rho,ENQ)
plt.title("Average number of waiting clients in M/M/1 system")
plt.xlabel("$\\rho$")
plt.ylabel("$\mathbf{E}[N_Q]$")
plt.show()
```

## Calculating E\[NQ\] for M/M/1/K+M model

```python
rho_nu=np.linspace(0.5,1.2,100)
ENQ_nu=[];

for r in rho_nu:
    l=r*100
    mu=100
    nu=1
    c=1
    K=100
    ENQ_nu.append(Erlang.ErwErlangC_ENQ(l,mu,nu,c,K))

plt.plot(rho,ENQ)
plt.plot(rho_nu,ENQ_nu)
plt.title("Average number of waiting clients in M/M/1/K+M system")
plt.xlabel("$\\rho$")
plt.ylabel("$\mathbf{E}[N_Q]$")
plt.ylim(0,25)
plt.legend(["M/M/1","M/M/1/K+M"])
plt.show()
```

## Economy of scale in M/M/c model

```python
rho=np.linspace(0.5,0.95,100)
cValues=[1,5,10]
ENQ=[];
for c in cValues:
    ENQ.append([])

for cIndex in range(len(ENQ)):
    for r in rho:
        c=cValues[cIndex]
        l=r*100*c
        mu=100
        ENQ[cIndex].append(Erlang.ErlangC_ENQ(l,mu,c))

for part in ENQ:
    plt.plot(rho,part)

plt.title("M/M/c systems with different c values")
plt.xlabel("$\\rho$")
plt.ylabel("$\mathbf{E}[N_Q]$")
plt.legend(cValues)
plt.show()
```
